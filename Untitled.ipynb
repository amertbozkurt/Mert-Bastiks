```python
import tensorflow as tf
import numpy as np
import numba as nb
from numba import njit,jit
import scipy as sci
import matplotlib as mpl
import matplotlib.pyplot as plt
import IPython.display as display
import PIL.Image

"""
This script is meant to read in a pre-trained neural network from those provided by Keras, extract a set of dense layers at different stages of the classification and repack them
as a new network, suitable to apply generated pictures to.
The effect that this hopefully achieves is that the artifacts of the different layers collected in the training process will be added to the injected picture. Amplifying these artifacts
is then expected to create gloomy atmospheres based on the input pictures.

"""
```

```python
"""
Here the pre-trained model is selected based on size, depth (best case is low with dense layers in different stages of the classification), and number of parameters.
The competition regarding performance is the need of dense structures while keeping the number of parameters reasonably low in order to minimize computational cost.

The choice is with DenseNet121
"""

pre_trained_model = tf.keras.applications.MobileNetV2(include_top=False)

pre_trained_model.summary()
```

```python
names = ['block_2_project','block_8_depthwise_BN'] #determines the layers of the underlying networks that are extracted

layers = [pre_trained_model.get_layer(name).output for name in names] # pulls the layers from the pre-trained network along the names provided above

arti_mod = tf.keras.Model(inputs = pre_trained_model.input,outputs = layers) #defines the reduced neural network used to collect artifacts
```

```python
def loss_func(inp,model):
    """
    Calculation of the comulatative loss of the layers used in the reduced model by propagating the input through the net and collecting the activations.
    
    
    inp: np.array(shape=(dim,dim,RGB)) where dim is picture dimension and RGB has size 3 for the distinct color channels
    
    model: tf.keras.model class 
    
    """
    activ = model(tf.expand_dims(inp,axis=0))
    if len(activ)==1:
        activ=[activ]
    
    vals = [tf.math.reduce_mean(acti) for acti in activ] 
    val = tf.reduce_sum(vals)
    """
    The definition here is, a.f.a.i.k. completely arbitrary redarding what measure is taking as loss
    """
    print(val)
    return val

```

```python
"""
In this cell the propagation network is defined as instance of tf.functions. 
To improve on overhead we perform the simplification:
    Rather than performing proper gradient ascent we only take two calls of the model
        and extrapolate linearly with a bound, randomized distance. 

"""

class Artifacter(tf.Module):
    def __init__(self,model):
        self.model = model
        
    @tf.function(
                jit_compile = True,
                input_signature = [tf.TensorSpec(shape=[None,None,3],dtype=tf.float32)]
                )
    def __call__(self,inp):
        
        loss = tf.constant(0.0)
        
        gradient = tf.gradients(loss_func(inp,self.model),inp)
        print(gradient)
        
        gradient /= tf.math.reduce_std(gradient)
        
        inp = inp +gradient*tf.random.uniform(shape=tf.shape(gradient),minval=1,maxval=5)
        inp = tf.math.tanh(inp) #normalize output
        
        return loss,inp
    
    
Artif = Artifacter(arti_mod) #this line instates the network
```

```python
def apply_sky_net(inp):
    inp = tf.keras.applications.mobilenet_v2.preprocess_input(inp) #this casts the input picture into a format that can be understood be the net
    inp = tf.convert_to_tensor(inp) #of course...
    
    los,img = Artif(inp)
    
    res = tf.cast(255*(img)/2,tf.uint8)
    
    return res
```
